```python
import gzip
import os

# 要压缩的 HTML 文件路径
html_file_path = '../index.html'
gzip_file_path = '../index.html.gz'

# 1. 压缩为 Gzip
with open(html_file_path, 'rb') as f_in:
    with gzip.open(gzip_file_path, 'wb') as f_out:
        f_out.writelines(f_in)

print(f"File compressed to {gzip_file_path}")

# 2. 读取 Gzip 文件并转换为十六进制
with open(gzip_file_path, 'rb') as f:
    gzip_content = f.read()

# 转换为十六进制格式并格式化为每行 16 个
hex_lines = []
for i in range(0, len(gzip_content), 16):
    line = ', '.join(f'0x{byte:02x}' for byte in gzip_content[i:i+16])
    # hex_lines.append(line)
    hex_lines.append(line + ',')  # 每行结尾加逗号

# 输出十六进制结果
for line in hex_lines:
    print(line)
```
